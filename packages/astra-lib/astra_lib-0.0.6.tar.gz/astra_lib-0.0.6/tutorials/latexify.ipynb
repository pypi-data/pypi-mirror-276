# Laltexify

- A guide to create research paper friendly plots using matplotlib.
- Checkout this [blog post](https://nipunbatra.github.io/blog/posts/2014-06-02-latexify.html) for motivation.

* Pre-requisites:
    - LaTeX should be installed on your system.

```python
import matplotlib.pyplot as plt
from astra.plot.utils import latexify
import seaborn as sns
```

## Line plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
```

### Vanilla plot

```python
plt.rcdefaults()

fig, ax = plt.subplots()
ax.plot(x, y, label='$y = x^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

### Latexified plot

```python
latexify(width=3, height=1.5, font_size=10)

fig, ax = plt.subplots()
ax.plot(x, y, label='$y = x^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
sns.despine(ax=ax)
fig.savefig('test.pdf') # Save in PDF format to preserve the quality
# OR if PDF consumes a lot of space
# fig.savefig('test.png', dpi=300) # Save in PNG format with 200-300 DPI
```

## How to add latexified plot in LaTeX document?

| Command | Correct/Incorrect | Reason |
| --- | --- | --- |
| `\includegraphics[width=0.5\textwidth]{figs/line_plot.pdf}` | Incorrect | We have already set the width of the plot in the code. |
| `\includegraphics[height=0.6\textwidth]{figs/line_plot.pdf}` | Incorrect | We have already set the height of the plot in the code. |
| `\includegraphics{figs/line_plot.pdf}` | Correct | We have already set the width and height of the plot in the code. |

## Exceptions

* Though latexify works well for most cases, in some cases, you might need to manually adjust the font sizes and override all the recommendations given in this guide. Feel free to do so for science!
