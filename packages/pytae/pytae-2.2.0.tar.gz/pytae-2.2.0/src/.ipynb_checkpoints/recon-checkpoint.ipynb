```python
import pandas as pd
import numpy as np
import pytae as pt

```

```python

df1=pd.DataFrame({'id':['a',None,'c','d','e','','f',np.NaN],
                  'balance':[10,20,np.NaN,21,15,10,20,25],
                  'country':['sg','cn','ca','np','in','in','np','in']})

df1
```

```python
df2=pd.DataFrame({'code':['c','f','d','d',np.NaN,'a','','g'],
                  'amt':[10,20,30,20,15,10,30,10],
                  'cntry':['uk','us','ca','np','in','jp','us','kr']})
df2
```

```python
k=pt.Recon(df1,{'id':'balance'},df2,{'code':'amt'})
```

```python
k.summary()
```

```python
k.values_recon()
```

```python
k.combined(count=True)
```

```python
k.values_recon(th=30,ignore_NA=False,agg=True)
```

```python
k.values_recon(th=0,ignore_NA=True)
```

```python
k.values_recon(th=0,ignore_NA=True,agg=True)
```
